## Position Performance Through the Years Cleaning

Creating a for loop to run through all of the files in the data folder, taking only rankings 1-8 (starting lineup), combining them, then adding a new column for the year to making plotting easier down the road.

```python
files = glob.glob('data/*csv')
dfs = []
for file in files:
    df = pd.read_csv(file)
    df['Year'] = file
    df = df[(df['Rk'] < 9)]
    dfs.append(df)
    df['Year'] = df['Year'].str.split('_').str[0]

df = pd.concat(dfs, ignore_index=True)
df['Year'] = df['Year'].str.split('/', n=1).str[-1]
df.to_csv('data/cleaned/batting_order_history.csv', index=False)
df
```
